```python
import pandas as pd

df = pd.read_csv('alice_data1.csv')
```

```python
def cleaner(string):
    string = string.replace(' | ВГТУ', '').replace('"', '').replace('-', ' ').replace(' -', '')

    if string[2:3] == '.' and string[5:6] == '.':
        string = string[9:]

    if '(' in string:
        for i in range(len(string)):
            if string[i] == '(': 
                indexes = i
                break
        string = string[:indexes - 1]

    return string.strip()

df['Название'] = [cleaner(df['Название'][i]) for i in df.index]
```

```python
display(df)
```

```python
unique_values = list(df['Название'].unique())

print(unique_values)

indexes = []

for value in unique_values:
    for i in df.index:
        if df['Название'][i] == value:
            indexes.append(i)
            break


print(len(unique_values), len(indexes))
```

```python
df = df.loc[indexes]

display(df)
```

```python
df.to_csv('data.csv', index=False)
```
